```python
X = df.iloc[: ,0:13]   # iloc mhanje location var jaun value ghya 
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split  # split kara data training and testing sathi
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)

# random state - parat data shuffle nahi hot
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()  # model cha ek object create kela
```

```python
lr.fit(X_train , y_train)  # training keli model chi
y_pred = lr.predict(X_test)  # testing keli
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error  # model kas kam kartay baghital
```

```python
print(f"MSE is : {mean_squared_error(y_test , y_pred)}")
print(f"MAE is : {mean_absolute_error(y_test , y_pred)}")
```
